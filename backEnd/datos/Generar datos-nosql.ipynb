## Crear archivos animales

```python
from datetime import datetime
import pandas as pd
import random as rd
import json
#ANIMALES
#MAX NUMBER OF ROWS
MAXNUM = 100
df = pd.read_csv('aac_shelter_cat_outcome_eng.csv')#.head(MAXNUM)
df = df[df['name'].notnull()]
df['date_of_birth'] = ((datetime.now() - pd.to_datetime(df['date_of_birth'])).dt.days/365).astype(int)
l = ['name','animal_type','breed','date_of_birth','sex','Spay/Neuter','color1']
print("Animales cargados....OK")
df[l].head()
```

```python
animal_columns = ["nombre","tipo","raza","edad","sexo","esterilizado","color"]
animal_df = df[l]
for i, column in enumerate(animal_columns):
    animal_df = animal_df.rename(columns={l[i]:animal_columns[i]})
animal_df.head()
```

```python
estadoL = ["adoptado","en_adopcion", "colonia", "perdido"]
```

```python
animal_df["estado"]  = animal_df["nombre"].apply(lambda x: rd.choice(estadoL))
animal_df = animal_df.head(100)
animal_df.head()
```

```python
animal_df = animal_df.reset_index().rename(columns={"index":"id"})
animal_df.head()
```

```python
animals_path = 'nosql_data/animal.csv'
animal_df.to_csv(animals_path,index=False,float_format=str)
pd.read_csv(animals_path).head(5)
```

## Formulario Poner en adopcion

```python
formulario_df = pd.read_pickle('formulario_adopciones').head(MAXNUM)
formulario_path = 'nosql_data/formulario.json'

formulario_df.head()
```

Como se cambio el numero maximo de personas entonces cambiar los duenos a duenos existentes, asimismo con los animales

```python
#available_peopleL = list
rd.choice(persona_df["cedula"])
rd.choice(animal_df.index)
print(persona_df.shape)
print(animal_df.shape)
pd.Series(animal_df.index)
```

```python
nforrows = formulario_df.shape[0]
formulario_df["duenio"] = persona_df.head(nforrows)["cedula"]#formulario_df["duenio"].apply(lambda x: rd.choice(persona_df["cedula"]))
formulario_df["animal"] = animal_df.head(nforrows)["id"]
formulario_df.head()
```

```python
def obtener_ushort(cedula):
    ushort_col = ["cedula","nombre","apellido","sexo"]
    resultado = persona_df[persona_df["cedula"]== cedula]
    ushort = {}
    for ucol in ushort_col:
        if resultado[ucol].values:
            ushort[ucol] = resultado[ucol].values[0] # ya se que solo va a ser 1
    return ushort
```

```python
formulario_df["usuario"] = formulario_df['duenio'].apply(obtener_ushort)
formulario_df.drop(columns=["duenio"], inplace = True)
formulario_df.head()
```

### Agregando animal 

```python
def obtener_ashort(pk):
    ushort_col = ["id","nombre","edad","raza","tipo"]
    #resultado = animal_df.iloc[pk]
    resultado = animal_df[animal_df["id"]== pk]
    ushort = {}
    for ucol in ushort_col:
        if resultado[ucol].values:
            ushort[ucol] = resultado[ucol].values[0] # ya se que solo va a ser 1
    return ushort
```

```python
formulario_df["animal"] = formulario_df['animal'].apply(obtener_ashort)
formulario_df.head()
```

```python
formulario_df.loc[0,"animal"]
```


### Guardando en json

```python
formulario_df.to_json(formulario_path, orient = 'records')
with open(formulario_path) as f:
    data = json.load(f)
#parsed = json.loads(formulario_path)
#json.dumps(parsed, indent=4)  
data
```
